#What is an imbalanced dataset? Describe two techniques to handle it theoretical + practical
#ANS

Oversampling: This technique involves increasing the number of instances in the minority class by duplicating them or creating synthetic samples using methods like SMOTE. It aims to balance the dataset by providing more examples of the minority class, which can help the model learn its characteristics better.
6
Undersampling: This technique reduces the number of instances in the majority class, which can help achieve balance but may lead to information loss if too many samples are removed. It is often used in combination with oversampling to maintain the integrity of the majority class while balancing the dataset.


These techniques are theoretical in nature, as they provide a conceptual understanding of how to handle imbalanced datasets. Practical implementations involve applying these techniques to real-world datasets and evaluating their effectiveness in improving model performance.


#Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.
#ANS
Absolute Maximum Scaling is a feature scaling method where each value is divided by the maximum absolute value of that feature. This transformation rescales the data so that values fall within the range of −1 to 1

Min-Max Scaling rescales features by subtracting the minimum value and dividing by the difference between the maximum and minimum values. This usually maps feature values to the range 0 to 1 while preserving the original distribution.

#Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
#ANS
In machine learning, categorical variables must be converted into numerical form for algorithms to process them effectively. Label Encoding and One-Hot Encoding are two widely used techniques for this transformation, each suited for different types of categorical data.

Label Encoding assigns each unique category an integer value. It’s efficient in terms of memory and works well for ordinal data where categories have a natural order

One-Hot Encoding creates a new binary column for each category, marking presence with 1 and absence with 0. It’s ideal for nominal data with no inherent order
